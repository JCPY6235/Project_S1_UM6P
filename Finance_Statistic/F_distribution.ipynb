```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

#### Fonction de densit√©

\frac{\sqrt{\frac{(d_1\,x)^{d_1}\,\,d_2^{d_2}} {(d_1\,x+d_2)^{d_1+d_2}}}} {x\,\mathrm{B}\!\left(\frac{d_1}{2},\frac{d_2}{2}\right)}\!

```python
rand_f_samples = stats.f.rvs(dfn=10, dfd=20, size=100000)

plt.figure(figsize=(10, 5))
plt.hist(
    rand_f_samples,
    density=True,
    color="lightgrey",
    edgecolor="darkgrey",
    bins="scott",
)


plt.plot(
    np.arange(0, 4, 0.1),
    stats.f.pdf(np.arange(0, 4, 0.1), dfn=10, dfd=20),
    "-",
    linewidth=2,
    color="orange",
)

plt.xlim(0, 4)

plt.title(
    "Histogram for a $F$-distribution with $v_1 = 10$ and $v_2 = 20$ degrees of freedom (df)"
)

plt.savefig("F_disctrib.png")

plt.show()
```

```python
x = 1.5
dfn = 10
dfd = 20

stats.f.pdf(x, dfn, dfd)  ## lower tail of the distribution --> [0,1.5]
```
