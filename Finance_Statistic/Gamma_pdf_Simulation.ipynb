```python
import numpy as np
import math as mt

def Binomial(N,p,k):
    C = mt.factorial(N)/(mt.factorial(k)*mt.factorial(N-k))
    return C * (p**k) * ((1-p)**(N-k))

P = np.array([Binomial(9,1/3,i) for i in range(5)]).sum()
P
```
